<b> Q3. The Job Role Columns holds the job information for each individual in this census snapshot. Using this column, create a Bar Chart that shows the count of 'Unique' Jobs per Job Group in the "Job-Role" Column in ascending order, as per the provided image below </b>


<b> Put your code below </b>

```python
# plot the bar chart using dataframe (after groupby and generate a new dataframe)
jlist=df.groupby('Job_Role')[' 77516'].count().sort_values()
jlist=pd.DataFrame(jlist)
jlist.rename(columns={' 77516':'Count'},inplace=True)
fig,ax= plt.subplots()
ax.bar(jlist.index,jlist['Count'])
ax.set_xticklabels(jlist.index,rotation=90)
ax.set_ylabel('Count of Different Job Roles')
plt.show()
```

```python
# after groupby generates a series with Job_role as indexes, and use series data to plot bar chart directly 
jr=df.groupby('Job_Role')['Job_Role'].count().sort_values()
fig,ax= plt.subplots()
ax.bar(jr.index,jr)
ax.set_xticklabels(jlist.index,rotation=90)
ax.set_ylabel('Count of Different Job Roles')
plt.show()
```

<b> Q5. Which Job Role has the highest <i> proportion </i> of individuals who earn >50K? </b>

<b> Put your code below </b>

```python
# create a series of data by grouping job_role and earnings, count the NO of entries
# create another series to count the total entries of each job role
je1=df.groupby(['Job_Role','Earnings'])[' 77516'].count()
je2=df.groupby('Job_Role')['Earnings'].count()

# get the percentage of two earnings for each job role 
je=je1/je2
df_je=pd.DataFrame(je)
df_je.reset_index(inplace=True,drop=False)
df_new=df_je.loc[df_je['Earnings']==' >50K']

# print the column names of the new dataframe
print(list(df_new.columns))

# plot the percentage data in a bar chart 
# confirm the Exec-mangerial job-role has the highest percentage of individuals who earn over 50k USD 
fig,ax=plt.subplots()
ax.bar(df_new['Job_Role'],df_new[0])
ax.set_xticklabels(df_new['Job_Role'],rotation=90)
ax.set_ylabel('the % of individuals make over 50k in each job role')
plt.show()
```
